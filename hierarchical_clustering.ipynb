```python
# Import Required Libraries
import pandas as pd
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt

# Load the Dataset
# Load the seeds dataset and separate grain varieties
seeds_df = pd.read_excel('/content/drive/MyDrive/AIML_datasets/seeds-less-rows.xlsx')
varieties = seeds_df.pop('grain_variety')  # Remove and save the grain varieties column

# Perform Hierarchical Clustering
mergings = linkage(seeds_df.values, method='complete')  # Complete linkage method

# Visualize the Dendrogram
plt.figure(figsize=(10, 6))
dendrogram(
    mergings,
    labels=varieties.tolist(),  # Use grain varieties as labels
    leaf_rotation=90,           # Rotate labels for better visibility
    leaf_font_size=6            # Set font size for labels
)
plt.title('Dendrogram of Grain Samples')
plt.xlabel('Grain Varieties')
plt.ylabel('Distance')
plt.show()

# Extract Flat Clusters
labels = fcluster(mergings, t=6, criterion='distance')  # Set max height for clustering

# Create DataFrame to Map Labels to Varieties
df = pd.DataFrame({'Cluster Label': labels, 'Grain Variety': varieties})
print(df.head())

# Cross-Tabulation to Compare Clusters with Varieties
ct = pd.crosstab(df['Cluster Label'], df['Grain Variety'])
print("\nCross-Tabulation:")
print(ct)
```

# Import Required Libraries

```python
import pandas as pd
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt
```

# Load the Dataset

```python
# Load the seeds dataset
seeds_df = pd.read_excel('/content/drive/MyDrive/AIML_datasets/seeds-less-rows.xlsx')

# Remove the grain species column (grain_variety) and save it for later
varieties = list(seeds_df.pop('grain_variety'))

# Extract the measurements as a NumPy array
samples = seeds_df.values
print(varieties)
seeds_df.head()
```

# Perform Hierarchical Clustering

```python
# Perform hierarchical clustering using the 'complete' linkage method
mergings = linkage(samples, method='complete')

# Create a dendrogram to visualize the hierarchical clustering
plt.figure(figsize=(10, 6))
dendrogram(
    mergings,
    labels=varieties,      # Use the grain varieties as labels
    leaf_rotation=90,      # Rotate the labels for better visibility
    leaf_font_size=6       # Set font size for labels
)
plt.title('Dendrogram of Grain Samples')
plt.xlabel('Grain Varieties')
plt.ylabel('Distance')
plt.show()
```

# Extract Flat Clusters

```python
# Extract flat clusters from the hierarchical clustering results
# Maximum height for clustering is set to 6
labels = fcluster(mergings, 6, criterion='distance')
```

# Create DataFrame with Labels and Varieties



```python
# Create a DataFrame to map cluster labels to the original grain varieties
df = pd.DataFrame({'labels': labels, 'varieties': varieties})

# Display the first few rows of the DataFrame
df.head()
```

# Cross-Tabulation

```python
# Perform a cross-tabulation to compare cluster labels with grain varieties
ct = pd.crosstab(df['labels'], df['varieties'])

# Display the cross-tabulation table
print(ct)
```
