# 2. Import Libraries

This cell sets up the necessary libraries for the project.

```python
import cv2
import numpy as np
```

# 3. Load the Pre-Trained Face Detection Model

Load the Haar Cascade model for frontal face detection.

```python
# Load the Haar Cascade for face detection
face_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/AIML_datasets/haarcascade_frontalface_default.xml')
```

# 4. Load the Image

Read the image file where you want to detect faces.

```python
# Read the image

import os
img = cv2.imread('/content/drive/MyDrive/AIML_datasets/people.jpg')

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

# 5. Detect Faces

Use the Haar Cascade to detect faces in the grayscale image.

```python
# Detect faces
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
print(f"Number of faces detected: {len(faces)}")
```

# 6. Draw Rectangles Around Detected Faces

Highlight detected faces in the original image.

```python
# Draw rectangles around faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
```

# 7. Display the Image
Show the image with rectangles drawn around detected faces.

```python
from google.colab.patches import cv2_imshow

# Display the result using cv2_imshow instead of cv2.imshow
cv2_imshow(img)  # This will show the image in Colab
cv2.waitKey(0)
cv2.destroyAllWindows()
```
