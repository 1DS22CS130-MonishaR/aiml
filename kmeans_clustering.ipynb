# What is K-Means Clustering?

K-Means clustering is an unsupervised machine learning algorithm used to group data into a predetermined number of `clusters (K)`. The algorithm partitions the dataset into K clusters such that the points in each cluster are as close as possible to the `cluster's centroid (mean)`. It minimizes the variance within each cluster and operates in the following steps:

1. **Initialization:** Select K initial cluster centroids, either randomly or based on specific initialization methods (e.g., k-means++).

2. **Assignment:** Assign each data point to the nearest cluster centroid based on the Euclidean distance.

3. **Update:** Recalculate the centroids of the clusters as the mean of all points assigned to them.

4. **Repeat:** Alternate between the assignment and update steps until convergence (i.e., centroids no longer move significantly or a maximum number of iterations is reached).

> **Key Characteristics:**

`K Value:` The number of clusters K is specified beforehand, often chosen based on domain knowledge or methods like the Elbow Method.

`Centroids:` These represent the center of each cluster, computed as the mean of all points in the cluster.

`Distance Metric:` Typically uses Euclidean distance, but other metrics can be applied depending on the use case.

> **Applications:**

K-Means is widely used in various domains, including:

1. Customer segmentation

2. Image compression

3. Anomaly detection

4. Organizing large datasets


> **Example:**

In this notebook, we apply K-Means clustering to a 2D dataset (ch1ex1.csv) and visualize the results. Each cluster is represented with a unique color, and the centroids are highlighted with markers. By varying the value of K, we can observe how the clustering changes.

```python
# Step 1: Import the necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 2: Load the dataset
# Replace 'ch1ex1.csv' with the correct path to your dataset if it's in a different directory
df = pd.read_excel('/content/drive/MyDrive/AIML_datasets/ch1ex1.xlsx')
print(df.head())
points = df.values
```

```python
# Step 3: Perform K-Means clustering with a specified number of clusters (K=3 in this case)
# Create and fit the model
model = KMeans(n_clusters=3, random_state=42)
model.fit(points)

# Predict the cluster labels for each data point
labels = model.predict(points)
labels
```

```python
# Step 4: Extract x and y coordinates from the dataset
xs = points[:, 0]  # Column 0 for x-coordinates
ys = points[:, 1]  # Column 1 for y-coordinates

# Visualize the clusters using a scatter plot
plt.scatter(xs, ys, c=labels)
plt.title("Scatter Plot of Points Colored by Cluster Labels")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Step 5: Get the coordinates of the cluster centroids
centroids = model.cluster_centers_
print(centroids)

# Separate x and y coordinates of the centroids
centroids_x = centroids[:, 0]
centroids_y = centroids[:, 1]
print(f"x:{centroids_x}, y:{centroids_y}")

# Plot the points and centroids together
plt.scatter(xs, ys, c=labels, label='Data Points')
plt.scatter(centroids_x, centroids_y, marker='X', s=100, c='red', label='Centroids')
plt.title("Scatter Plot with Cluster Centroids")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```
