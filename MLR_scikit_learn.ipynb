# **Multiple linear regression model** for housing_prices dataset and predict house price based on the area, floor and room size of the house using the library **scikit learn**.
- Find out the accuracy of the model using R2 score statistics for the predicted model.

```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading dataset
df=pd.read_csv("/content/drive/MyDrive/AIML_datasets/housing_prices.csv")
df.head()
```

```python
# Setting Target and Feature Vectors
x = df.iloc[:, :3].values  # Features (AREA, FLOOR, ROOM)
y = df.iloc[:, 4].values   # Target (PRICE)
```

```python
# Splitting the dataset
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
# Fitting the model
from sklearn.linear_model import LinearRegression

mlr_model = LinearRegression(fit_intercept=True)
mlr_model.fit(x_train, y_train)

# Displaying the intercept and coefficients
print(mlr_model.intercept_)  # Intercept (b0)
print(mlr_model.coef_)       # Coefficients [b1, b2, b3]
```

```python
# Finding R2 score
print(mlr_model.score(x_train, y_train))  # R2 score for training set
print(mlr_model.score(x_test, y_test))    # R2 score for test set
```
